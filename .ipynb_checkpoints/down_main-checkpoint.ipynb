```python
import scipy
import vqe_methods
import pyscf_helper

import pyscf
from pyscf import lib
from pyscf import gto, scf, mcscf, fci, ao2mo, lo, molden, cc
from pyscf.cc import ccsd

import openfermion as of
from openfermion import *
from tVQE import *

import numpy as np
```

```python
geometry = [('Be', (0,0,0))]
n_act = 5
n_frz = 0
act_max = n_frz+n_act
shift = 2*n_frz
charge = 0
spin = 0
basis = 'cc-pvdz'

[n_orb, n_a, n_b, h, g, mol, E_nuc, E_scf, C, S] = pyscf_helper.init(geometry,charge,spin,basis,reference='rhf')
sq_ham = pyscf_helper.SQ_Hamiltonian()
sq_ham.init(h, g, C, S)
print(" HF Energy: %21.15f" %(E_nuc + sq_ham.energy_of_determinant(range(n_a),range(n_b))))
```

```python
mf = scf.RHF(mol)
mf.conv_tol_grad = 1e-14
mf.max_cycle = 1000
mf.verbose = 4
mf.init_guess = 'atom'
mf.kernel()
print("MO energies:")
print(mf.mo_energy)

mccsd = cc.RCCSD(mf)
mccsd.conv_tol = 1e-12
mccsd.conv_tol_normt = 1e-10
#myccsd.diis_start_cycle=10000
mccsd.max_cycle = 1000
mccsd.kernel()
t1_amps = mccsd.t1
t2_amps = mccsd.t2

print(t1_amps.shape)
print(t2_amps.shape)
```

```python
fmat = sq_ham.make_f(range(n_a),range(n_b))
vmat = sq_ham.make_v()
print(fmat.shape)
print(vmat.shape)
print(n_orb)
print(n_a)
print(n_b)
```

```python
# getting t1_external               
for i in range(0,n_a):
    for a in range(n_a,act_max):
        t1_amps[i,a-n_a]=0

print(t1_amps.shape)
#print(t1_amps)


# getting t2_external
for i in range(0,n_a):
    for j in range(0,n_a):
        for a in range(n_a,act_max):
            for b in range(n_a,act_max):
                t2_amps[i,j,a-n_a,b-n_a]=0
                
print(t2_amps.shape)
#print(t2_amps)
```

```python
# expanding the t1 amplitude into alpha and beta space
t1 = np.zeros((4,24))
for i in range(0,n_a):
    for a in range(n_a,n_orb):
        ia = 2*i
        ib = 2*i + 1
        aa = 2*a
        ab = 2*a + 1
        t1[ia,aa-(n_a+n_b)] = t1_amps[i,a-n_a]
        t1[ib,ab-(n_a+n_b)] = t1_amps[i,a-n_a]

print(t1.shape)

# expanding the t2 amplitude into alpha and beta space
t2 = np.zeros((4,4,24,24))
for i in range(0,n_a):
    ia = 2*i
    ib = 2*i+1
    for j in range(0,n_a):
        ja = 2*j
        jb = 2*j+1
        for a in range(n_a,n_orb):
            aa = 2*a
            ab = 2*a+1
            for b in range(n_a,n_orb):
                ba = 2*b
                bb = 2*b+1 
                t2[ia,ja,aa-(n_a+n_b),ba-(n_a+n_b)]=t2_amps[i,j,a-n_a,b-n_a]
                t2[ib,jb,ab-(n_a+n_b),bb-(n_a+n_b)]=t2_amps[i,j,a-n_a,b-n_a]

print(t2.shape)
```
